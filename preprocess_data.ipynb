```python
def safe_to_datetime(series):
    return pd.to_datetime(series, errors='coerce', dayfirst=True, infer_datetime_format=True)

print(df['Data da Coleta'].dtype)
print(df['Data dos primeiros sintomas'].dtype)
```

```python
df['Data da Coleta'] = safe_to_datetime(df['Data da Coleta'])
df['Data dos primeiros sintomas'] = safe_to_datetime(df['Data dos primeiros sintomas'])
print(df['Data da Coleta'].dtype)
print(df['Data dos primeiros sintomas'].dtype)
print(df['Data da Coleta'].unique())
```

```python
df_filtro_intervalo_primeiros_sintomas = df.copy()
# Aplicar a função para cada coluna de data
# (df_filtro_intervalo_primeiros_sintomas['Data da Coleta'])
# 1. Remover linhas com valores vazios na coluna 'dataInicioSintomas'
df_filtro_intervalo_primeiros_sintomas = df_filtro_intervalo_primeiros_sintomas.dropna(subset=['Data dos primeiros sintomas'])
# 2. Calcular as datas mínima e máxima dos exames para cada linha
filtro_intervalo = ((df_filtro_intervalo_primeiros_sintomas['Data da Coleta'] >=
df_filtro_intervalo_primeiros_sintomas['Data dos primeiros sintomas'] + pd.Timedelta(days=1)) &
(df_filtro_intervalo_primeiros_sintomas['Data da Coleta'] <= df_filtro_intervalo_primeiros_sintomas['Data dos primeiros sintomas'] + pd.Timedelta(days=15)))

df_filtro_intervalo_primeiros_sintomas = df_filtro_intervalo_primeiros_sintomas[filtro_intervalo]
df = df_filtro_intervalo_primeiros_sintomas.copy()
```

```python
# Função para corrigir os nomes na coluna "codigoLaboratorioPrimeiraDose"
def corrigir_laboratorio(laboratorio):
    correcoes = {
    'FUNDACAO OSWALDO CRUZ': 'ASTRAZENECA',
    'ASTRAZENECA/FIOCRUZ': 'ASTRAZENECA',
    '085 - COVID-19 ASTRAZENECA/FIOCRUZ - COVISHIELD' : 'ASTRAZENECA',
    'ASTRAZENECA/OXFORD': 'ASTRAZENECA',
    '089 - COVID-19 ASTRAZENECA - ChAdOx1-S' : 'ASTRAZENECA',
    'SERUM INSTITUTE OF INDIA LTD': 'ASTRAZENECA',
    'BioNTech/Fosun Pharma/Pfizer': 'PFIZER',
    '087 - COVID-19 PFIZER - COMIRNATY' : 'PFIZER',
    '099 - COVID-19 PEDIÁTRICA - PFIZER COMIRNATY' : 'PFIZER',
    '102 - COVID-19 PFIZER - COMIRNATY PEDIÁTRICA MENOR DE 5 ANOS' :
    'PFIZER',
    '103 - COVID-19 PFIZER - COMIRNATY BIVALENTE' : 'PFIZER',
    '3 - Comirnaty (Pfizer)' : 'PFIZER',
    'Pfizer/BioNTech' : 'PFIZER',
    'PFIZER/BIONTECH': 'PFIZER',
    'PFIZER MANUFACTURING BELGIUM NV BELGICA' : 'PFIZER',
    'PFIZER - PEDIÁTRICA' : 'PFIZER',
    'PFIZER - PEDIÁTRICA MENOR DE 5 ANOS' : 'PFIZER',
    'WYETH PHARMACEUTICALS' : 'PFIZER',
    '086 - COVID-19 SINOVAC/BUTANTAN - CORONAVAC' : 'CORONAVAC',
    '1 - Coronavac - Sinovac (Butantan)' : 'CORONAVAC',
    'BUTANTAN' : 'CORONAVAC',
    'INSTITUTO BUTANTAN' : 'CORONAVAC',
    'FUNDACAO BUTANTAN' : 'CORONAVAC',
    'Sinovac' : 'CORONAVAC',
    'SINOVAC LIFE SCIENCE CO LTD' : 'CORONAVAC',
    'SINOVAC': 'CORONAVAC',
    'SINOVAC/BUTANTAN': 'CORONAVAC',
    'MINISTERIO DA SAUDE': 'Pendente Identificacao',
    '088 - COVID-19 JANSSEN - Ad26.COV2.S' : 'JANSSEN',
    '000 - ~Ignorado/Não se aplica/Não informado' : 'Pendente Identificacao',
    'Pendente Identifica??o' : 'Pendente Identificacao'
    }
    return correcoes.get(laboratorio, laboratorio)
# Aplica a função corrigir_laboratorio às colunas "laboratorioPrimeiraDose" e "laboratorioSegundaDose"
df_vacinas_corrigidas = df.copy()
df_vacinas_corrigidas['Vacina aplicada COVID-19 1ª dose'] = df_vacinas_corrigidas['Vacina aplicada COVID-19 1ª dose'].apply(corrigir_laboratorio)
df_vacinas_corrigidas['Vacina aplicada COVID-19 2ª dose'] = df_vacinas_corrigidas['Vacina aplicada COVID-19 2ª dose'].apply(corrigir_laboratorio)
df_vacinas_corrigidas['Vacina aplicada COVID-19 dose de reforço'] = df_vacinas_corrigidas['Vacina aplicada COVID-19 dose de reforço'].apply(corrigir_laboratorio)
```

```python
#Classificar vacina pela tecnologia usada
df_classifica_vacina_por_tecnologia = df_vacinas_corrigidas.copy()
def classificar_vacina(nome):
    if nome in ["ASTRAZENECA", "JANSSEN"]:
        return "vetor viral"
    elif nome == "CORONAVAC":
        return "vírus inativado"
    elif nome == "PFIZER":
        return "rna mensageiro"
    else:
        return "Pendente Identificacao" # Para valores desconhecidos
# Criar a nova coluna baseada na classificação
df_classifica_vacina_por_tecnologia["Tecnologia utilizada 1 dose"] = df_classifica_vacina_por_tecnologia["Vacina aplicada COVID-19 1ª dose"].apply(classificar_vacina)
df_classifica_vacina_por_tecnologia["Tecnologia utilizada 2 dose"] = df_classifica_vacina_por_tecnologia["Vacina aplicada COVID-19 2ª dose"].apply(classificar_vacina)
df_classifica_vacina_por_tecnologia["Tecnologia utilizada dose reforco"] = df_classifica_vacina_por_tecnologia["Vacina aplicada COVID-19 dose de reforço"].apply(classificar_vacina)
```

```python
df['Vacina rna mensageiro 1 dose'] = df['Tecnologia utilizada 1 dose'].apply(lambda x: 1 if 'rna mensageiro' in x else 0)
df['Vacina vetor viral 1 dose'] = df['Tecnologia utilizada 1 dose'].apply(lambda x: 1 if 'vetor viral' in x else 0)
df['Vacina vírus inativado 1 dose'] = df['Tecnologia utilizada 1 dose'].apply(lambda x: 1 if 'vírus inativado' in x else 0)
df['Vacina rna mensageiro 2 dose'] = df['Tecnologia utilizada 2 dose'].apply(lambda x: 1 if 'rna mensageiro' in x else 0)
df['Vacina vetor viral 2 dose'] = df['Tecnologia utilizada 2 dose'].apply(lambda x: 1 if 'vetor viral' in x else 0)
df['Vacina vírus inativado 2 dose'] = df['Tecnologia utilizada 2 dose'].apply(lambda x: 1 if 'vírus inativado' in x else 0)
df['Vacina rna mensageiro dose reforco'] = df['Tecnologia utilizada dose reforco'].apply(lambda x: 1 if 'rna mensageiro' in x else 0)
df['Vacina vetor viral dose reforco'] = df['Tecnologia utilizada dose reforco'].apply(lambda x: 1 if 'vetor viral' in x else 0)
df['Vacina vírus inativado dose reforco'] = df['Tecnologia utilizada dose reforco'].apply(lambda x: 1 if 'vírus inativado' in x else 0)
df = df.drop(['Tecnologia utilizada 1 dose', 'Tecnologia utilizada 2 dose', 'Tecnologia utilizada dose reforco', 'Vacina aplicada COVID-19 1ª dose', 'Vacina aplicada COVID-19 2ª dose', 'Vacina aplicada COVID-19 dose de reforço'], axis=1)
```
